## Step 3:
Create a function to get news and calculate the sentiment

````python
# We use TextBlob() to calculate sentiment - show students an example:

print(TextBlob("This stock is amazing!").sentiment.polarity)  # ➝ ~0.8 to 1.0 (positive)
print(TextBlob("This stock is okay.").sentiment.polarity)     # ➝ ~0.2 (slightly positive)
print(TextBlob("This stock is bad.").sentiment.polarity)      # ➝ ~-0.7 (negative)
print(TextBlob("This is a stock.").sentiment.polarity)    
````

````python
# Function to get the average sentiment of a stock based on news articles from a specific day
def get_sentiment(stock, date):
    # Convert the input 'date' to string format 'YYYY-MM-DD' for the API
    from_date = date.strftime('%Y-%m-%d')
    # Calculate the 'to_date' as the next day to cover the entire 24-hour period
    to_date = (date + timedelta(days=1)).strftime('%Y-%m-%d')

    # Define the parameters required by the API for querying news articles
    params = {
        'q': stock,                    # Search query term (the stock name or symbol)
        'from': from_date,              # Start date for articles
        'to': to_date,                  # End date for articles
        'sortBy': 'relevancy',          # Sort articles by how relevant they are to the query
        'apiKey': API_KEY,              # API key for authentication (assumed to be defined earlier)
        'language': 'en',               # Only retrieve English articles
        'pageSize': 5                   # Limit to 5 articles to control API usage and response size
    }

    # Call the API using the requests library with the specified parameters
    response = requests.get(BASE_URL, params=params)
    # Convert the API response to JSON format for easier data access
    data = response.json()
    
    # Debug print to inspect the raw API response (optional but useful for troubleshooting)
    print(data)

    sentiments = []  # Initialize an empty list to store sentiment scores for each article
    
    # Iterate over each article in the API response (if 'articles' exists, else returns empty list)
    for article in data.get('articles', []):
        # Extract the title and description from each article; use empty string as default if missing
        title = article.get('title', '')
        description = article.get('description', '')
        
        # If both title and description are missing, skip this article
        if not title and not description:
            continue
        
        # Combine the title and description text for sentiment analysis
        text = f"{title} {description}"
        # Use TextBlob to calculate the sentiment polarity of the text 
        # (returns a value between -1 [negative] and 1 [positive])
        sentiment = TextBlob(text).sentiment.polarity
        sentiments.append(sentiment)  # Store the sentiment score

    # If any sentiments were collected, calculate and return their mean (average)
    if sentiments:
        mean = np.mean(sentiments)  # Calculate the average sentiment score
        print(mean)  # Optional debug print to show the calculated mean
        return mean  # Return the average sentiment for that day

    # If no sentiments were found, return None (indicating no relevant articles found)
    return None
````

## Step 4:
Simulate Backtesting

````python
def gather_daily_sentiments(stocks, start_date, end_date):
    """
    For each date in [start_date, end_date], calls get_sentiment(stock, date)
    and returns a DataFrame of shape (n_dates × n_stocks) of daily polarity.
    """
    dates = pd.date_range(start_date, end_date)  # Generate a range of dates between start_date and end_date
    # Initialize an empty DataFrame with dates as rows and stocks as columns to store sentiment scores
    sentiments = pd.DataFrame(index=dates, columns=stocks, dtype=float)

    # Loop through each date in the date range
    for date in dates:
        # For each stock, calculate and store the sentiment score for the given date
        for stock in stocks:
            sentiments.at[date, stock] = get_sentiment(stock, date)
    return sentiments  # Return the completed DataFrame of sentiment scores


def fetch_price_data(stocks, start_date, end_date):
    """
    Downloads daily OHLC data for each ticker in `stocks` between start_date and end_date.
    Returns a dict mapping ticker → its DataFrame.
    """
    price_data = {}  # Initialize an empty dictionary to store price DataFrames per stock

    # Loop through each stock ticker and download its price data
    for stock in stocks:
        df = yf.download(
            stock,
            start=start_date.strftime('%Y-%m-%d'),  # Format start_date as a string for the API call
            end=(end_date + timedelta(days=1)).strftime('%Y-%m-%d'),  # Include end_date by adding 1 day
            interval='1d',  # Fetch daily data (Open, High, Low, Close)
            progress=False  # Disable the progress bar
        )
        price_data[stock] = df[['Open', 'Close']]  # Store only 'Open' and 'Close' columns for each stock

    return price_data  # Return the dictionary mapping each stock to its price DataFrame
````

````python
# Define parameters for the API and functions
START_DATE = pd.to_datetime('2025-04-28')
END_DATE   = pd.to_datetime('2025-05-09')
STOCK_UNIVERSE = ['AAPL','DNUT','BRK-B']

# 1) get your sentiment surface
sentiment_df = gather_daily_sentiments(STOCK_UNIVERSE, START_DATE, END_DATE)

# 2) pull in all your price data
price_dict   = fetch_price_data(STOCK_UNIVERSE, START_DATE, END_DATE)

# now students will have have:
# - sentiment_df.loc[date, stock] → polarity of the stock on that date
# - price_dict[stock].loc[date, ['Open','Close']] → prices
````
